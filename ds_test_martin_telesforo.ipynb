Now, I seek for the original transactions checking every column match except by date.  The purchase date need to be before of the return.

```python

dfs = []
for i in range(0,len(pct_negatives.index)):
    dfs.append(pct.loc[(pct['TRANS DATE'] <pct_negatives['TRANS DATE'].iloc[i] ) &
                       (pct['ORIGINAL GROSS AMT'] ==pct_negatives['ORIGINAL GROSS AMT'].iloc[i] ) &
                       (pct['MERCHANT NAME'] ==pct_negatives['MERCHANT NAME'].iloc[i] ) &
                       (pct['CARD NUMBER'] ==pct_negatives['CARD NUMBER'].iloc[i] ) &
                       (pct['ORIGINAL CUR'] ==pct_negatives['ORIGINAL CUR'].iloc[i] ) &
                       (pct['DIRECTORATE'] ==pct_negatives['DIRECTORATE'].iloc[i] ) &
                       (pct['TRANS CAC CODE 1'] ==pct_negatives['TRANS CAC CODE 1'].iloc[i] ) &
                       (pct['TRANS CAC CODE 2'] ==pct_negatives['TRANS CAC CODE 2'].iloc[i] ) &
                       (pct['BILLING CUR CODE'] ==pct_negatives['BILLING CUR CODE'].iloc[i] )
        ] )
```

AS the most of variables are categorical and we have an imbalanced data the best fast approach is execute a Desicion Tree model.

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

I split the data and prepared a Grid Search, to ensure same proportion of responders I use the stratify option.


```python
X = pct_refounds.drop(['REFOUND','TRANS DATE'] , axis=1)
y = pct_refounds['REFOUND']
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size=0.30, random_state=8)
```

```python
X_train["ORIGINAL CUR"] = X_train["ORIGINAL CUR"].astype("category")
X_train["BILLING CUR CODE"] = X_train["BILLING CUR CODE"].astype("category")
X_train["MERCHANT NAME"] = X_train["MERCHANT NAME"].astype("category")
X_train["DIRECTORATE"] = X_train["DIRECTORATE"].astype("category")
X_train["TRANS CAC CODE 1"] = X_train["TRANS CAC CODE 1"].astype("category")
X_train["TRANS CAC CODE 2"] = X_train["TRANS CAC CODE 2"].astype("category")
X_train["CARD NUMBER"] = X_train["CARD NUMBER"].astype("category")

X_test["ORIGINAL CUR"] = X_test["ORIGINAL CUR"].astype("category")
X_test["BILLING CUR CODE"] = X_test["BILLING CUR CODE"].astype("category")
X_test["MERCHANT NAME"] = X_test["MERCHANT NAME"].astype("category")
X_test["DIRECTORATE"] = X_test["DIRECTORATE"].astype("category")
X_test["TRANS CAC CODE 1"] = X_test["TRANS CAC CODE 1"].astype("category")
X_test["TRANS CAC CODE 2"] = X_test["TRANS CAC CODE 2"].astype("category")
X_test["CARD NUMBER"] = X_test["CARD NUMBER"].astype("category")
```

```python
model = XGBClassifier(   tree_method='hist', enable_categorical=True,objective='binary:logistic')
```

I'm not using a big list of parameters but here is a opportunity to improve the model.

```python
params = {'min_child_weight': [1, 5],'gamma': [0.5, 1, 1.5], 'subsample': [0.6, 0.8],'colsample_bytree': [0.6, 0.8],'max_depth': [3, 4],'n_estimators' : [600],'learning_rate' : [0.05,0.1]}
```

```python
fkfold=StratifiedKFold( shuffle = True, random_state = 8)
grid_serach = RandomizedSearchCV(model , param_distributions=params, n_iter=7, scoring='roc_auc',  cv=fkfold.split(X_train,y_train), random_state=8 )
```

```python
grid_serach.fit(X_train, y_train)
```

```python
print(grid_serach.best_estimator_)
```

```python
grid_serach.best_params_
```

Lets see the performance of the "Best" model get it

```python
y_val = grid_serach.predict_proba(X_test)
```

As predicted when I got the 614 responders,  the ROC score is very bad comparing directly with the probability output and y_test.

```python

false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_val[:,1])
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

```python

plt.figure(figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate,true_positive_rate, color='red',label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```

Now lets check the  confusion matrix taking the cut off the minimum of the probability of the top decile in the Train data. But the plot of above is advicing a bad behavior.

```python
y_train_pred = grid_serach.predict_proba(X_train)
y_trainp_df=pd.DataFrame( y_train_pred[:,1],columns=['Predicted_y'])
```

```python
df2 = y_trainp_df[y_trainp_df['Predicted_y'] > y_trainp_df['Predicted_y'].quantile(0.9)]
min_cutoff=df2.min()
y_pred = np.where(y_val[:,1] >min_cutoff[0], 1, 0)
```

```python
confusion_matrix=metrics.confusion_matrix(y_test,y_pred)
confusion_matrix
```

```python
metrics.ConfusionMatrixDisplay (confusion_matrix).plot()
```
